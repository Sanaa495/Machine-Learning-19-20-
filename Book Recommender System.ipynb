```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
users=pd.read_csv('BX-Users.csv', sep=';', encoding='latin-1', error_bad_lines=False)
books=pd.read_csv('BX-Books.csv', sep=';', encoding='latin-1', error_bad_lines=False)
ratings=pd.read_csv('BX-Book-Ratings.csv', sep=';', encoding='latin-1', error_bad_lines=False)
```

```python
print(users.shape)
print(books.shape)
print(ratings.shape)
```

```python
ratings.rename(columns={'User-ID':'user_id','Book-Rating':'score'},inplace=True)
```

```python
ratings.head()
```

```python
x=ratings['user_id'].value_counts()>200
padhaku=x[x].index.tolist()
padhaku
```

```python
ratings=ratings[ratings['user_id'].isin(padhaku)]
ratings.shape
```

```python
ratings['ISBN'].nunique()
```

```python
books=books[['ISBN','Book-Title','Book-Author','Year-Of-Publication','Publisher']]
```

```python
books_with_rating=ratings.merge(books,on='ISBN')
```

```python
a=books_with_rating['ISBN'].value_counts()>50
famous=a[a].index.tolist()
len(famous)
```

```python
books_with_rating=books_with_rating[books_with_rating['ISBN'].isin(famous)]
```

```python
books_with_rating.shape
```

```python
books_with_rating.drop_duplicates(subset=['user_id','Book-Title'],inplace=True)
```

```python
books_with_rating.shape
```

```python
pt=books_with_rating.pivot_table(index='Book-Title', columns='user_id',values='score')
```

```python
pt.shape
```

```python
pt.fillna(-1,inplace=True)
```

```python
pt
```

```python
from scipy.sparse import csr_matrix
matrix=csr_matrix(pt)
```

```python
matrix
```

```python
from sklearn.neighbors import NearestNeighbors
NN=NearestNeighbors(algorithm='brute')
```

```python
pt.shape
```
